```python
import pandas as pd
```

```python
data = pd.read_csv('./input.txt',sep='\t',header=-1)
data.columns = ['time','content']
data.head()
```

```python
stopwords = set()
with open('./Chinese-StopWords.txt') as f:
    for i in f.readlines():
        stopwords.add(i.strip('\n'))
stopwords
```

```python
data['remove_stop'] = data.content.apply(lambda s: ' '.join([x for x in s.split(' ') if x not in stopwords]))
data.head()
```

```python
data[data.content != data.remove_stop]
```

```python
from sklearn import feature_extraction
```

```python
tfidf = feature_extraction.text.TfidfVectorizer()
tf_idf = tfidf.fit_transform(data.content)
```

```python
tf_idf.shape, data.shape
```

```python
import hdbscan
```

```python
cls = hdbscan.HDBSCAN(min_samples=1)
```

```python
cls.fit_predict(tf_idf)
```

```python
from sklearn import manifold
```

```python
projector = manifold.t_sne.TSNE()
```

```python
tsen_res = projector.fit_transform(tf_idf.toarray())
```
